## Datos categoricos

### Resultados de una encuesta sobre si volveran a visitar un sitio turistico (si:s

```r
encuesta <-c("si", "no", "no", "si", "ns/nc", "si", "no", "si", "ns/nc", "no")
```

### Creacion de una tabla de frecuencias absolutas:

```r
table(encuesta)
```

### Frecuencias relativas:

```r
table(encuesta)/length(encuesta)
```

## Datos Numericos

###  Grado de satisfaccion de un cliente 

```r
satisfaccion <- c(1, 3, 2, 5, 1, 2, 3, 4, 2, 3, 4, 3, 5, 3, 4, 5, 2, 3, 4, 2)
```

### Tabla de Frecuencias:

```r
table(satisfaccion)
```

### Diagrama de barras:

```r
barplot(table(satisfaccion), xlab="satisfaccion",ylab="frecuencia")
```

### Diagrama de sectores:

```r
pie(table(satisfaccion),main="Satisfacción")
```

### Dıas de baja laboral durante el ultimo año de los miembros de un departamento de produccion:

```r
bajas <- c(2, 3, 12, 23, 0, 4, 12, 4, 13, 2, 0, 2, 1, 4, 0, 2)
```

```r
table(bajas)
```

### Diagrama de tallo y hojas:

```r
stem(bajas)
```

### Media, mediana y moda:

```r
mean(bajas)
```

```r
median(bajas)
```

```r
which.max(table(bajas))
```

### Añadir un dato y ver como afecta a la media y mediana

```r
bajas <- c(bajas, 150)
```

```r
mean(bajas)
```

```r
median(bajas)
```

```r
which.max(table(bajas))
```

## Medidas de dispersión

### Varianza, desviacion tıpica y percentiles:

```r
var(bajas)
```

```r
sd(bajas)
```

```r
quantile(bajas,.5)
```

```r
quantile(bajas,c(.25,.5,.75))
```

### Resumen de datos numericos:

```r
summary(bajas)
```

### Rango interquartilico

```r
IQR(bajas)
```
