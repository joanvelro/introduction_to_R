## Asimetrıa

Las medidas de asimetría, sobre todo el coeficiente de asimetría de Fisher, junto con las medidas de apuntamiento o curtosis se utilizan para contrastar si se puede aceptar que una distribución estadística sigue la distribución normal. Esto es necesario para realizar numerosos contrastes estadísticos en la teoría de inferencia estadística.

[Coeficiente de Asimetría de Fisher](https://es.wikipedia.org/wiki/Asimetr%C3%ADa_estad%C3%ADstica)

$$ \gamma_1 = \frac{\mu_3}{\sigma_3} = \frac{\sum_i (x_i - \hat{x})^3}{n\cdot s^3}$$

donde s es la [desviacion estandar](https://es.wikipedia.org/wiki/Desviación_t%C3%ADpica)

* El valor de la asimetrıa es 0 para una distribucion simetrica (normal). 
* Cola a la izquierda, asimetrıa< 0.
* Cola a la derecha, asimetrıa> 0.

```r
asimetria = sum((espera-mean(espera))^3)/ (length(espera)*sqrt(var(espera))^3)
asimetria
if (asimetria<0){
    print("Asimetria a la izquierda")
    }
if (asimetria>0){
    print("Asimetria a la derecha")
    } 
if (asimetria==0){
    print("Datos simetricos")
    }
```

```r
asimetria <- function(x){
    sum((x-mean(x))^3)/ (length(x)*sqrt(var(x))^3) 
}
```

## Curtosis o Apuntamiento

una medida que sirve para analizar el grado de concentración que presentan los valores 
de una variable analizada alrededor de la zona central de la
distribución de frecuencias, sin necesidad de generar el gráfico.


[Coeficiente de apuntamiento](https://es.wikipedia.org/wiki/Curtosis)

$$ \beta_2 = \frac{\mu_4}{\sigma_4} = \frac{\sum(x_i - \hat{x})^4}{n \cdot s^4} - 3 $$

* El valor de curtosis para una distribucion normal es 0. 
* Platicutica, curtosis < 0
* Mesocurtica, curtosis ≈ 0
* Leptocurtica, curtosis > 0


```r
apuntamiento = sum((espera-mean(espera))^4)/ (length(espera)*var(espera)^2)-3
apuntamiento
if (apuntamiento<0){
    print("Curtosis: Platicutica")
    }
if (apuntamiento>0){
    print("Curtosis: Leptocurtica")
    } 
if (apuntamiento<0.1 && apuntamiento>-0.1){
    print("Curtosis: Mesocurtica")
    }
if (apuntamiento==0){
    print("Curtosis nula")
    }
```

```r
apuntamiento <- function(x){
   sum((x-mean(x))^4)/ (length(x)*var(x)^2)-3  
} 
```


## [Boxpot](https://en.wikipedia.org/wiki/Box_plot)

* Ponen en una sola dimensión los datos de un histograma
* Proporcionan una visión general de la simetría de la distribución de los datos
* Si la mediana no está en el centro del rectángulo, la distribución no es simétrica.
* Son útiles para ver la presencia de valores atípicos también llamados outliers.
* El 50% de la población está en los límites de la caja.
* Permite ver como es la dispersión de los puntos con la mediana, los percentiles 25 y 75 y los valores máximos y mínimos.


<img src="boxplot.png" width="50%">

### formado por cinco puntos: lower, upper, q1, q2 y q3 

```r
q1 = quantile(bajas,.25)
q1
```

```r
q2 = quantile(bajas,.50)
q2
```

```r
q3 = quantile(bajas,.75)
q3
```

```r
iqr <- IQR(espera)
```

```r
lower = min(c(q1-1.5*iqr,q3+1.5*iqr))
lower
```

```r
upper = max(c(q1-1.5*iqr,q3+1.5*iqr))
upper
```
